````python
# Dataset Duplicate Value Count
duplicate_rows= df[df.duplicated(keep=False)]
duplicate_rows
````

````python
# Checking for duplicate values and found null after successfully removing duplicates. Now Data Is Duplicate Free
duplicate_rows= df[df.duplicated(keep=False)]
duplicate_rows
````

#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
round(((df.isnull().sum()/len(df.index))*100),2)
````

````python
# Since country columnn had .52% null values, so these rows have been removed
df = df.dropna(subset=['country'])
````

````python
# Normalization of agent column by median
# calculating median of agent column
norm_data= df[(df['agent']>=df['agent'].quantile(.25)) & (df['agent']<=df['agent'].quantile(.75))]
````

````python
norm_data['agent'].median()
````

````python
# filling up blank agent rows with its median value
df['agent'] = df['agent'].fillna(norm_data['agent'].median())
````

````python
df.info()
````

````python
# Replacing all null values with nan in the dataset
df=df.fillna(value=pd.np.nan)
````

````python
# Visualizing the missing values
# we can well see all null values replaced with Nan
df.head()
````
